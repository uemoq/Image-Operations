## Horizontal Flip

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as image
im = image.imread("parrots.jpg")
im = im[:, ::-1, :]
# Alternatively use: im = np.fliplr(im)
plt.imshow(im)
plt.show()
```

## Downsample Images

```python
image.thumbnail("parrots.jpg", "thumbnail.png", 0.1)
im = image.imread("thumbnail.png")
plt.imshow(im, interpolation="nearest")
plt.show()
```
